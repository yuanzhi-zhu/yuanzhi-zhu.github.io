## The first step of this ipynb is to reveal a locked ZhiHu website(a sensitive answer or personal particle)

````python
import json
import requests as rq
from bs4 import BeautifulSoup
````

This url is a blocked particle, you can not see the content of it if you open it directly in you browser

````python
# headers are required to successfully connect to the ZhiHu sever
headers = {
        'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.88 Safari/537.36',
        'Content-Type': 'application/x-www-form-urlencoded',
        'Connection' : 'Keep-Alive',
        'Accept':'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8'
}

url = '\
https://zhuanlan.zhihu.com/p/27659587'
````

````python
BlockPage=rq.get(url,headers=headers)
````

Woops, the real data is hiden in the javascript, and we need first try to show this kind of content
<font color=blue>By comparing the blocked and unblocked ZhiHu website, we can find that the original code block: </font>
```javascript

<font color=blue>is replaced by:</font>

```
And we can see from the source code that the original content is coded in the format of JSON. Then what we do is to convert this part back to HTML language. (requests and BeautifulSoup do not support javascript content)

````python
script = BlockPage_soup.find_all('script')[2].string
#data = script.split("bootstrapData['menuMonthWeeks'] = ", 1)[-1].rsplit(';', 1)[0]
data = json.loads(script)
#data
````

````python
data
````

After getting the above result, by manually testing we can find that the text in the label 'content' is the original content of the particle.

Then we need to process this 'data' varible.

````python
type(data)
````

To make the following code generalized, we need to get the articles id"

````python
article_id=url.rsplit('/',1)[1]
original_header = '<div class="RichText ztext Post-RichText"> '
original_tail = ' <\div>'
original_HTML_content=data['initialState']['entities']['articles'][article_id]['content']
original_HTML_content = original_header+original_HTML_content + original_tail
original_content_soup=BeautifulSoup(original_HTML_content, "lxml")
#print(original_content_soup.prettify())
original_content_tag = original_content_soup.find_all('div')[0]
print(original_content_tag.prettify())
````

Afte the first test, I find that the picture can't be loaded correctly, because the 'src' value of the 'img' tag is modified when we convert the json code. The following redirects the 'src' to the original or actual picture, you can choose original or actual picture using the varible 'original'

````python
original=True

for figure in original_content_tag.find_all('figure'):
    if original:
        figure.find_all('img')[1]['src']=figure.find_all('img')[1]['data-original']
    else:
        figure.find_all('img')[1]['src']=figure.find_all('img')[1]['data-actualsrc']
````

Next step is putting this part back to replace the trash part.

Luckly, we can use BeautifulSoup to edit the tag(both name and content):
<font color=green> find out the useless tag and replace it with the tag containing original content of the article</font>

````python
trash_content = BlockPage_soup.find_all('div',{'class':'PostIndex-warning'})[0]
#trash_content
print(trash_content.prettify())
````

````python
trash_content.replace_with(original_content_tag)
trash_content
````
